```python
def min_max_norm(df):
    df_2 = df.copy()

    width = df_2['width'].to_numpy()[:, np.newaxis].astype('float32')
    height = df_2['height'].to_numpy()[:, np.newaxis].astype('float32')
    # width = df_2['width'].to_numpy()[:5, np.newaxis]
    # height = df_2['height'].to_numpy()[:5, np.newaxis]

    # df_3 = df_2.to_numpy()[:5, 2:8]
    df_3 = df_2.to_numpy().astype('float32')

    # df_3[:, 2:-2:2]   = df_3[:, 2:-2:2] / width
    # df_3[:, 3:-2:2]   = df_3[:, 3:-2:2] / height
    df_3[:, :-2:2]   = df_3[:, :-2:2] / np.concatenate((width[:879], width[:879]), axis=0)
    df_3[:, 1:-2:2]   = df_3[:, 1:-2:2] / np.concatenate((height[:879], height[:879]), axis=0)

    return width, height, df_3
```

```python
temp_1 = ((normalized_df_2 - center_coord_2)**2).astype('float32')
temp_2 = temp_1[:, ::2] + temp_1[:, 1::2]
temp_3 = np.nanmean(np.sqrt(temp_2), axis=1)
print(len(temp_3))
```

```python
temp_3
```

```python
def scaling_factor(df):
    # return 1/df
    # return np.sqrt(2)/df
    return df/np.sqrt(2)
```

```python
s_factor = scaling_factor(temp_3)

# print(s_factor)
print(len(s_factor))
```

```python
def matrix_calc(s_fact, x, y, x_mean, y_mean):
    mat_calc = np.array([[(1/s_fact), 0, -1*(1/s_fact)*x_mean], [0, (1/s_fact), -1*(1/s_fact)*y_mean], [0, 0, 1]])
    # mat_calc = np.array([[s_fact, 0, -1*s_fact*x_mean], [0, s_fact, -1*s_fact*y_mean], [0, 0, 1]])
    output = (mat_calc @ (np.array([x, y, 1]).T))

    return mat_calc, output, output.T[0], output.T[1]
```

```python
mat_calc, output, x_t, y_t = matrix_calc(s_factor[1], normalized_df_2[0][2], normalized_df_2[0][3],\
                                         center_coord_2[0][2], center_coord_2[0][3])

print(x_t, y_t)
```

```python
# 원래 좌표대로 복구
from numpy.linalg import inv

def matrix_calc_inv(matrix_calc, matrix_out):
    matrix_calc_inv = inv(matrix_calc)

    output_2 = matrix_calc_inv @ matrix_out

    return output_2[0], output_2[1]
```

```python
# 1269, 827
matrix_calc_inv(mat_calc, output)
```

```python
def convert_df(df, mean_df, s_val):
    df_2 = df.copy()
    output_arr  = []
    mat         = []

    for i in range(len(df_2)):
        for j in range(0, len(df_2.T)-1, 2):
            mat_calc, output, x_t, y_t = matrix_calc(s_factor[i], df_2[i, j], df_2[i, j+1], mean_df[i, j], mean_df[i, j+1])

            df_2[i, j]      = x_t
            df_2[i, j+1]    = y_t

            output_arr.append(output)
            mat.append(mat_calc)

    return df_2, np.array(output_arr), np.array(mat)
```
